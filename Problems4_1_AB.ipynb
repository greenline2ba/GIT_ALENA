```python
for item in df['Item']:
    print(item)
```

```python
milk = df[df['Item'] == 'Milk']
print(milk)
```

```python
goatsmilk = df[(df['Item'] == 'Milk') & (df['Unit Price ($)'] == 3.99)]
print(goatsmilk)
```

```python
pennyitems = df[(df['Unit Price ($)']< 1)]
print(pennyitems)
                
```

```python
df['Celkova cena'] = df['Quantity'] * df['Unit Price ($)']
df
```

```python
print(df['Celkova cena'].sum())
print(df['Celkova cena'].mean())
```

```python
cohojenajviac = df.groupby('Item')['Unit Price ($)'].mean()
cohojenajviac
```

```python
sale = df[df['Type'] == 'Sale']
print(sale)
grouping = df.groupby('Item')['Quantity'].sum()
print(grouping)
```
