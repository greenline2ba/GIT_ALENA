```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("Finance_Trends.csv")
df
```

```python
source = df.groupby('gender').size()
print(source)
```

```python
fig = px.pie(
    source,
    values=source.values,
    names=source.index,)
```

```python
fig.show()
```

```python
fig = px.bar(
    source,
    x ='Gender',
    y ='Age',
    )
```
